```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.tree import DecisionTreeClassifier
import pickle
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
# CSV dosyasını yükleme   #yeni
data = pd.read_csv('files/duygusal_durumlar.csv')
```

```python
# Metin sütununu alın    #yeni
corpus = data['text'].tolist()
```

```python
# Etiket sütununu alın    #yeni
labels = data['tag'].tolist()
```

```python
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(corpus)
```

```python
print(X)
```

```python

veri_seti_sayisi = data.shape[0]

print("Veri setinin satır sayısı:", veri_seti_sayisi)
```

```python
tag_sayisi = data['tag'].value_counts()

# Tag değeri 0 olanların sayısı
tag_0_sayisi = tag_sayisi[0]

# Tag değeri 1 olanların sayısı
tag_1_sayisi = tag_sayisi[1]

# Tag değeri 2 olanların sayısı
tag_2_sayisi = tag_sayisi[2]

print("Tag 0 olanların sayısı:", tag_0_sayisi)
print("Tag 1 olanların sayısı:", tag_1_sayisi)
print("Tag 2 olanların sayısı:", tag_2_sayisi)
```

```python
form = pd.read_csv('files/form-added-new.csv')
```

```python
# Satır sayısını al
num_rows = form.shape[0]
num_cols = form.shape[1]
my_cols = num_cols-4
print("Column Nums: "+str(num_cols))
print("Row Nums: "+ str(num_rows))
```

```python
# Kullanıcıları döngü ile işleme
meta_data = {} # Boş sözlük oluştur
for i in range(num_rows):
    # Kullanıcı verisini al
    user = form.to_numpy()[i][5:]

    # ID numarasını oluştur
    id_num = i + 1

    # Kullanıcıyı sözlüğe ekle
    meta_data[id_num] = user
    
```

```python
def sentiment_find(sum):
    if sum < 0 or sum > 46:
        return "Geçersiz değer"
    elif sum <= 5:
        return "Depresyon yok"
    elif sum < 10:
        return "Normal depresyon"
    elif sum <= 18:
        return "Hafif depresyon"
    elif sum <= 30:
        return "Orta depresyon"
    elif sum <= 35:
        return "Ağır depresyon"
    else:
        return "Aşırı depresyon"
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, labels, test_size=0.3, random_state=5)
```

```python
model_dt = DecisionTreeClassifier()
```

```python
model_dt.fit(X_train, y_train)
```

```python
y_pred = model_dt.predict(X_test)
```

```python
testsonuclari = []
```

```python
accuracy_dt = accuracy_score(y_test, y_pred)
print("DESICION TREE -- Test doğruluk oranı:", accuracy_dt)
```

```python
testsonuclari.append(accuracy_dt)
```

```python
from math import nan
import string  # Noktalama işaretlerini kaldırmak için string modülünü kullanacağız

score = {}
id_num = 1

for i in range(num_rows):
    user = meta_data[i + 1]
    
    scores = []  # Kullanıcıya ait skor listesi
    for j in range(min(my_cols, len(user))):
        a = user[j]
        if not pd.isna(a):  # pd.isna() ile NaN değerlerini kontrol edin
            if isinstance(a, float):
                a = a
            else:
                # Noktalama işaretlerini kaldır
                a = a.translate(str.maketrans('', '', string.punctuation))
                a = a.lower()  # Cümleyi küçük harfe çevir
                # Vektörleştirme
                X = loaded_vectorizer.transform([a])
                # Tahmin yapma
                print(a)
                y_pred = loaded_model.predict(X)
                scores.append(y_pred[0])  # Skor listesine ekleme
                print(y_pred)
    print(f"{'*'*10}")
    score[id_num] = scores  # Her kullanıcı için ayrı skor listesi ekleme
    id_num += 1
    print(id_num)
    
print(f"{'*' * 20}")
print(score)
```

```python
score_totals = {}    
for key, value in score.items():
    print(key)
    print(value)
    total_score = sum(value) 
    diagnosis = sentiment_find(total_score)  # Sentiment değerini hesaplama
    score_totals[key] = {
        "Total Score": total_score,
        "Diagnosis": diagnosis,
    }
    print(score_totals)
    print(f"*"*5)
```

```python
for key, value in score_totals.items():
    print(f"ID: {key} - Total Score: {value['Total Score']} - diagnosis: {value['Diagnosis']}")
```

```python
df = pd.DataFrame.from_dict(score_totals, orient='index', columns=['Total Score', 'Diagnosis'])
```

```python
df.to_csv('files/score_totals_decion-tree.csv', index_label='Id')
```

## GradientBoostingRegressor----------------------------------------------------

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_squared_error
```

```python
data = pd.read_csv('files/duygusal_durumlar.csv')
corpus = data['text'].tolist()
target = data['tag'].tolist()
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(corpus)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, target, test_size=0.3, random_state=5)
```

```python
# GradientBoostingRegressor modelini kullanacağız
model = GradientBoostingRegressor(n_estimators=75, learning_rate=0.4, random_state=5)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
accuracy_gdr= 1-mse
print("1 - Ortalama Kare Hata (MSE):", accuracy_gdr)
```

```python
testsonuclari.append(accuracy_gdr)
```
